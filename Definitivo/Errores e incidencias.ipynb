```python
balances_y_transacciones=transacciones.merge(balances, on='t_id', how='inner')
```

```python
balances_y_transacciones.head()
```

```python
balances_y_transacciones.drop(['incremento_balance_or'], axis=1)
```

```python
#Crear nueva columna con incrementos de balances
balances_y_transacciones["incremento_balance_or"] = balances_y_transacciones["balance_prev_or"] - balances_y_transacciones["balance_post_or"]
```

```python
balances_y_transacciones["incremento_balance_des"] = balances_y_transacciones["balance_post_des"] - balances_y_transacciones["balance_prev_des"]
```

```python
balances_y_transacciones.head()
```

```python
# Comprobar filas en que coincidan los incrementos de los balances de los clientes origen con las cuantías de las transacciones
balances_y_transacciones[balances_y_transacciones['incremento_balance_or'] == balances_y_transacciones['monto']]
```

```python
# Comprobar filas en que coincidan los incrementos de los balances de los clientes origen con las cuantías de las transacciones
balances_y_transacciones[balances_y_transacciones['incremento_balance_or'] != balances_y_transacciones['monto']]
```
